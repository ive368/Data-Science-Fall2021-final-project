```python
import csv
import numpy as np
import os
import subprocess

def fileclean(prefix):
    try:
        os.remove(prefix + ".ins")
    except FileNotFoundError:
        pass
    try:
        os.remove(prefix + ".pcf")
    except FileNotFoundError:
        pass
    try:
        os.remove(prefix + ".prp")
    except FileNotFoundError:
        pass

def printOutput(file_prefix):
    f = open(file_prefix + '.prp', 'r')
    file_content = f.read()
    print(file_content)
    return f.close()
```

```python
from pynput.keyboard import Key, Controller
import time
keyboard = Controller()

def typekeyinp(file):
    time.sleep(2)
    return keyboard.type(file)
stringdict={}
def typesplitter(string):
    stringfrag=string.splitlines()
    for s in range(len(stringfrag)):
        #print(s)
        stringdict[s]=stringfrag[s]
    return stringdict
    
    

def typekeyinp1(file):
    typesplitter(file)
    time.sleep(5)
    for n in stringdict:
        keyboard.type(stringdict[n])
        time.sleep(0.25)
        keyboard.press(Key.enter)
        time.sleep(0.25)
    return
```
